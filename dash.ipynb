```python
from dotenv import load_dotenv
load_dotenv()

import os
from qdrant_client import QdrantClient

client = QdrantClient(url=os.environ["QDRANT_URL"], api_key=os.environ.get("QDRANT_APIKEY"))
col = os.environ["QDRANT_COLLECTION"]

print("collections:", [c.name for c in client.get_collections().collections])
info = client.get_collection(col)
print("collection ok:", info.config.params)
```
